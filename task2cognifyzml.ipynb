```python
import pandas as pd

df = pd.read_csv('/content/sample_data/Dataset .csv', encoding='ISO-8859-1')

def recommend(cuisine, price):
    result = df[(df['Cuisines'].str.contains(cuisine, na=False)) & (df['Price range'] == price)]
    return result[['Restaurant Name', 'Cuisines', 'Aggregate rating']].head(5)

print(recommend("Indian", 2))
```
